STEP 1


import pandas as pd

df = pd.read_csv("Downloads/Titanic-Dataset.csv")
summary_stats = df.describe()
print(summary_stats)


STEP 2

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12, 6))
df[['Age', 'Fare']].hist(bins=20, edgecolor='black', figsize=(12, 6))
plt.tight_layout()
plt.show()
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[['Age', 'Fare']])
plt.title('Boxplots of Age and Fare')
plt.show()

```

STEP 3

```python

sns.pairplot(df[['Age', 'Fare', 'Pclass', 'SibSp', 'Parch']])
plt.show()
corr_matrix = df[['Age', 'Fare', 'Pclass', 'SibSp', 'Parch']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

STEP 4

```python
fig, ax = plt.subplots(1, 2, figsize=(14, 5))

sns.histplot(df['Age'].dropna(), bins=30, kde=True, ax=ax[0])
ax[0].set_title('Age Distribution')

sns.histplot(df['Fare'], bins=30, kde=True, ax=ax[1])
ax[1].set_title('Fare Distribution')

plt.tight_layout()
plt.show()
```

STEP 5


```python
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1

sns.countplot(x='FamilySize', data=df)
plt.title('Family Size Distribution')
plt.show()
df['AgeGroup'] = pd.cut(df['Age'], bins=[0, 12, 18, 35, 60, 100], 
                        labels=['Child', 'Teen', 'Young Adult', 'Adult', 'Senior'])
sns.countplot(x='AgeGroup', hue='Survived', data=df)
plt.title('Survival by Age Group')
plt.show()
```
